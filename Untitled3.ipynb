```python
import numpy as np  # For numerical operations
import pandas as pd  # For data manipulation
import tensorflow as tf  # Deep learning framework
from tensorflow.keras.models import Sequential  # To define the model
from tensorflow.keras.layers import Dense, Dropout  # Layers for MLP
from tensorflow.keras.optimizers import SGD, Adam, RMSprop  # Optimizers
from tensorflow.keras.regularizers import l2  # For weight decay (regularization)
from tensorflow.keras.callbacks import LearningRateScheduler, ReduceLROnPlateau  # Learning rate scheduling
from sklearn.model_selection import train_test_split  # For dataset splitting
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt  # For plotting results
```

```python
df = pd.read_csv('/content/train.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop('id', axis=1) # Drop the 'id' column
df.head()
```

```python
df.describe()
```

```python
X = df.drop('species', axis=1)
y = df['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2025)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=2025)
```

```python
X.shape, y.shape
```

```python
df['species'].nunique()
```

```python
# Initialize LabelEncoder
le = LabelEncoder()

# Fit and transform the target variable
y_train = le.fit_transform(y_train)
y_val = le.transform(y_val)
y_test = le.transform(y_test)

# Now y_train, y_val, and y_test contain numerical labels
y_train
```

```python
y.nunique()
```

```python
# Define the model
model = Sequential([
    # Input layer with size matching the flattened image dimensions
    Dense(units=128, input_shape=(X.shape[1],), activation='tanh', name='Hidden_Layer'),
    # Hidden layer with tanh activation
    Dense(units=y.nunique(), activation='softmax', name='Output_Layer')
])

# Compile the model
model.compile(
    optimizer='adam',  # You can experiment with optimizers
    loss='sparse_categorical_crossentropy',  # Use this loss for integer labels
    metrics=['accuracy']  # Track accuracy
)

# Summary
model.summary()
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_val, y_val))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
# Plot training & validation accuracy values
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
from tensorflow.keras.callbacks import LearningRateScheduler
from tensorflow.keras.optimizers import Adam, SGD, RMSprop

def build_model(hidden_size, num_classes, input_size):
    """Builds a 3-layer MLP model."""
    model = Sequential([
        Dense(units=hidden_size, input_shape=(input_size,), activation='tanh', name='Hidden_Layer'),
        Dense(units=num_classes, activation='softmax', name='Output_Layer')
    ])
    return model

def train_model(X_train, y_train, X_val, y_val, batch_size, hidden_size, learning_rate, optimizer_choice, epochs,num_classes=99,input_size=X.shape[0]):
    """Trains the MLP model with specified hyperparameters."""
    # Build the model
    num_classes = len(set(y_train))  # Determine number of classes from the data
    input_size = X_train.shape[1]  # Input size matches flattened image size
    model = build_model(hidden_size, num_classes, input_size)
    # Compile the model
    # Choose optimizer
    if optimizer_choice == 'adam':
        optimizer = Adam(learning_rate=learning_rate)
    elif optimizer_choice == 'sgd':
        optimizer = SGD(learning_rate=learning_rate)
    elif optimizer_choice == 'rmsprop':
        optimizer = RMSprop(learning_rate=learning_rate)
    else:
        raise ValueError("Invalid optimizer choice. Use 'adam', 'sgd', or 'rmsprop'.")



    # Compile the model
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Define a learning rate scheduler
    def lr_schedule(epoch):
        return learning_rate * (0.1 ** (epoch // 10))  # Decrease LR every 10 epochs

    lr_scheduler = LearningRateScheduler(lr_schedule)

    # Train the model
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        batch_size=batch_size,
        epochs=epochs,
        callbacks=[lr_scheduler],
        verbose=2
    )

    return model, history

```

```python
# Example Usage
# Assuming X_train, y_train, X_val, and y_val are preprocessed datasets
batch_size = 32
hidden_size = 128
learning_rate = 0.001
optimizer_choice = 'adam'
epochs = 100

model, history = train_model(
    X_train, y_train, X_val, y_val,
    batch_size=batch_size,
    hidden_size=hidden_size,
    learning_rate=learning_rate,
    optimizer_choice=optimizer_choice,
    epochs=epochs
)
```

```python
# Plot training & validation accuracy values
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```
